## 결측치 처리
- **fillna**  
DataFrame.**fillna(value = None, method = None, axis = None, inplace = False)**

> value : 결측 데이터에 채울 값 입력. default(None)  

> method : 결측 데이터를 채우는 방식을 지정. default(None)
>    > bfill/backfill : 이후 데이터로 채움  
>    > pad/fifill : 이전 데이터로 채움  

> axis : 데이터의 축을 지정. default(None)  
>    > 0 or "index" : 가로 기준 축(Row별)  
>    > 1 or :columns" : 세로 기준 축(column별)  
> inplace : 처리 후 기존 데이터 수정. default(False)

## Dummy 변수 생성
- **get_dummies**  
DataFrame.**get_dummies(data, prefix = None, prefix_sep = '_' ) 

> data : 더미변수 생성 데이터. default(None)  
> prefix : 생성될 더미변수의 접두어. 예) prefix = "var" -> var1, var2, ... default(None)  
> prefix_sep : 접두어 뒤의 구분자. 예) prefix_sep = "_ " -> var_1, var_2. default(_)

## 데이터 분할
- **train_test_split**  
sklearn.model_selection.**train_test_split(test_size, shuffle, random_state)**  

> test_size : test 데이터 크기 지정. default(None, 0.25). 0 ~ 1.0  
> train_size : train 데이터 크기 지정. default(None, test size와 동일). 0 ~ 1.0  
> shuffle : 데이터 분리 전 데이터 혼합(섞음) 여부 지정. default(True)  
> random_state j: 초기 자료 선택 기준. 값에 따라 선택되는 데이터가 달라짐. default(None)

## 모델 훈련
- **fit**  
적용모델.fit(X,Y)
> : 정의된 모델에 훈련용(train) 데이터(X,Y)를 입력해 훈련된 모델을 생성  
> X, Y : 훈련용(train) X 및 Y 데이터

## 모델 적용
- **predict**  
> ; fit 적용으로 생성된 모델에 평가/시험용(test) 데이터(X)를 입력해 Y 예측 및 모델의 성능평가  
> X : 평가 또는 시험용(test) X 데이터. 결과로 예측된 Y 데이터 제공  

## 결과 예측
- **score**  
> : fit 적용으로 생성된 모델에 새로운 데이터(X,Y는 옵션)을 입력해 Y 예측(soring 점수화)  
> X,Y : 예측용(score) X (또는 Y) 데이터

## 모델 최적 파라미터 탐색
- **GridSearchCV**  
sklearn.model_selectioin.**GridSearchCV(estimatoor, param_grid, scoring, ...)**

> estimator : 적용 모델 지정

> param_grid : 탐색 대상 파라미터별 입력 값 지정. 적용 모델별 파라미터 지정  

> scoring : 모델 평가 기준 지정  
>    > accuracy : (분류) 모델의 정분류율  
>    > f1 : (분류) 정밀도(precision)와 재현율(recall) 지수를 이용한 정확도 지수
>    > ... 다양한 분류 모델의 성능 평가 기준 제공  
>    > r2 : (예측) 모델의 설명력  
>    > ... 다양한 예측 모델의 성능 평가 기준 제공 

> n_jobs : 병렬 처리를 위한 사용 core 수 지정. default(1). -1 : 모든 core 사용

## 상관계수 계산
- **pandas.DataFrame.corr**  
pandas.DataFrame.**corr(method = "pearson", min_periods = 1)**
> mothod : 상관계수 계산 기준. default(pearson)  
>    > pearson : Pearson 상관계수 (연속성)  
>    > kendall : Kendal Tau 상관계수 (이산형)
>    > spearman : Spearman 순위(rank, 비모수) 상관계수  

> min_periods : 상관관계 산출을 위한 최소 데이터 수. default(1)

## 회귀분석
- **OLS**  
statsmodels.regression.linear_model.**OLS(endog, exog = None, missing = "none")**  
> endog(array-like) : 목표변수(endogenous)  
> exog(array-like) : 설명변수(exogenous)  
> missing(str) : 결측치 처리 옵션. default(None)  
>    > none : 결측치 확인하지 않음  
>    > drop : 결측치가 있는 경우 제외  
>    > raise : 결측치가 있는 경우 Error 발생
